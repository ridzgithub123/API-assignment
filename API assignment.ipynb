```python
#Q2. Give advantages and disadvantages of using API.
'''
Advantages:

Increased Interoperability: APIs enable different software systems to communicate with each other,
regardless of their programming language or platform. 
This enhances the interoperability between systems, allowing for the exchange of data and functionality .

Modularity: APIs bring a new level of modularity to applications.
Developers can leverage the expertise of other applications when developing an application, which means they no longer need to
reinvent the wheel when it comes to things like authentication, communication, payment processing, and maps .

Flexibility: APIs provide flexibility in terms of how data is accessed and used. 
They allow developers to access data from multiple sources and use it in different ways.

Scalability: APIs can be scaled up or down depending on the needs of the application.
This means that as an application grows, it can continue to use the same API without having to change its code .

Disadvantages:

Security Risks: APIs can be a security risk if not implemented properly. 
They can be vulnerable to attacks such as SQL injection, cross-site scripting (XSS), and cross-site request forgery (CSRF).

Dependency: Applications that rely heavily on APIs are dependent on the availability and reliability of those APIs.
If an API goes down or is unavailable, it can cause problems for the application that relies on it .

Complexity: APIs can be complex to implement and maintain. 
They require a certain level of technical expertise to use effectively .

Costs: Some APIs may require payment for usage beyond a certain threshold .

'''
```

```python
#Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.
'''
REST :

REST stands for REpresentational State Transfer.
It is a software architectural style that defines a set of constraints to be used for creating web services .
RESTful web services are those that follow the REST architectural style

The REST architecture is based on a client-server model where the client requests resources from the server,
and the server responds with the requested resources . The communication between the client and server happens
through the Internet’s Hypertext Transfer Protocol (HTTP) .

There are six architectural constraints that make any web service RESTful.

1) Uniform Interface: It suggests that there should be a uniform way of interacting with a given server
irrespective of device or type of application (website, mobile app) .
2)Stateless: It means that the necessary state to handle the request is contained within the request itself,
and the server would not store anything related to the session . 
In REST, the client must include all information for the server to fulfill the 
request whether as part of query params, headers, or URI 1. Statelessness enables greater
availability since the server does not have to maintain, update or communicate that session state.

3)Cacheable: Every response should include whether the response is cacheable or not and
for how much duration responses can be cached at the client-side .
The client will return data from its cache for any subsequent request,
and there would be no need to send the request again to the server.

4)Client-Server: The client and server should be separate from each other and 
should interact through HTTP requests and responses.

5)Layered System: A layered system is one in which different components are responsible for different functions .

6)Code on Demand (Optional): This constraint allows clients to download code from servers
and execute it in their own environment.


SOAP :
The SOAP specifications describe a distributed processing model in which a SOAP messageis passed between SOAP nodes.
The message originates at a SOAP senderand is sent to a SOAP receiver.
Between the sender and the receiver, the message might be processed by one or more SOAP intermediaries.


## SHORT COMINGS OF SOAP:

REST is an architecture that’s more data-driven, while SOAP is a standardized protocol for transferring structured information that’s more function-driven. 
REST permits many different data formats, including plain text, HTML, XML, and JSON, which is a great fit for data and yields more browser compatibility; 
SOAP only uses XML
'''
```
